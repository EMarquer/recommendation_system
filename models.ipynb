## Anon

```python
def anonymous_recommandation(df, n = 20):
    count = df[['id','reviews.rating','city']].groupby(['id']).count().values    
    
    hotels =df[['id','name','reviews.rating']].groupby(['id']).mean()
    hotels['count'] = count
    hotels['rate_count'] = hotels['count'] * hotels['reviews.rating']
    
    return hotels.nlargest(n, 'rate_count' )
```

```python
anonymous_recommandation(df)
```
